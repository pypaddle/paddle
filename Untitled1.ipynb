# Imports

```python
import os
import paddle
import torchvision
import torch
import torch.nn as nn
import numpy as np
import paddle.sparse
import seaborn as sns
import networkx as nx
import copy

from paddle.learning import train, test
from torchvision.transforms import transforms
```

# Dataset: MNIST

```python
data_base_path = '/home/julian/data'

batch_size = 100

custom_transform = transforms.Compose([
    transforms.ToTensor(),  # first, convert image to PyTorch tensor
    transforms.Normalize((0.1307,), (0.3081,))  # normalize inputs
])
# download and transform train dataset
mnist_train_loader = torch.utils.data.DataLoader(torchvision.datasets.MNIST(os.path.join(data_base_path, 'set/mnist/'),
                                                          download=True,
                                                          train=True,
                                                          transform=custom_transform),
                                                           batch_size=batch_size,
                                                           shuffle=True)

# download and transform test dataset
mnist_test_loader = torch.utils.data.DataLoader(torchvision.datasets.MNIST(os.path.join(data_base_path, 'set/mnist/'),
                                                         download=True,
                                                         train=False,
                                                         transform=custom_transform),
                                                          batch_size=batch_size,
                                                          shuffle=True)
```

```python
model = paddle.sparse.MaskedDeepFFN(784, 10, [200, 100])
optimizer = torch.optim.SGD(model.parameters(), lr=0.01, momentum=0.5)
loss_func = nn.CrossEntropyLoss()
device = torch.device('cuda:0' if torch.cuda.is_available() else 'cpu')
```

```python
initial_training_epochs = 100
print('Training ..')
for epoch in range(initial_training_epochs):
    train(mnist_train_loader, model, optimizer, loss_func, device, percentage=True)

print('Trained.')
model.eval()
print('Test:', test(mnist_test_loader, model, device))
```

```python
model_copy = copy.deepcopy(model)
model_copy.eval()
print('Test:', test(mnist_test_loader, model_copy, device))
```

```python
model_copy.eval()
print('Test:', test(mnist_test_loader, model_copy, device))
```

```python
model_copy.train()
model_copy_optimizer = torch.optim.SGD(model_copy.parameters(), lr=0.01, momentum=0.5)
print('Train:', train(mnist_train_loader, model_copy, model_copy_optimizer, loss_func, device, percentage=True))
model_copy.eval()
print('Test:', test(mnist_test_loader, model_copy, device))
```
